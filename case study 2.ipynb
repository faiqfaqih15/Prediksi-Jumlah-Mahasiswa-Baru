### **REGRESI LINEAR**

```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

data = pd.read_csv('DATA MAHASISWA DAN UANG PANGKAL v4.csv')
```

### **TREND JUMLAH MAHASISWA**

```python
plt.figure(figsize=(10, 6))
plt.plot(data['TAHUN'], data['JUMLAH MAHASISWA'], marker='o')
plt.title('Tren Jumlah Mahasiswa FTI (2013-2022)')
plt.xlabel('TAHUN')
plt.ylabel('JUMLAH MAHASISWA')
plt.grid(True)
plt.show()
```

### **TREND UANG PANGKAL**

```python
plt.figure(figsize=(10, 6))
plt.plot(data['TAHUN'], data['UANG PANGKAL'], marker='o')
plt.title('Tren UANG PANGKAL FTI (2013-2022)')
plt.xlabel('TAHUN')
plt.ylabel('UANG PANGKAL')
plt.grid(True)
plt.show()
```

```python
X_train = data[['TAHUN', 'UANG PANGKAL']].values
y_train = data['JUMLAH MAHASISWA'].values

model = LinearRegression()
model.fit(X_train, y_train)
UangPangkalTerakhir = data['UANG PANGKAL'].iloc[-1]

#uang pangkal naik 10%
Xtest10Persen = [[2023, 1.1 * UangPangkalTerakhir]]
ypred10Persen = model.predict(Xtest10Persen)
print("=====================================================================")
print("Uang pangkal naik 10% dari uang pangkal tahun 2022: Rp.",int(1.1 * UangPangkalTerakhir))
print("Prediksi Jumlah Mahasiswa di jika Uang Pangkal naik 10%:", int(ypred10Persen[0]))
print("=====================================================================")
print()

#uang pangkal 20%
Xtest20Persen = [[2023, 1.2 * UangPangkalTerakhir]]
ypred20Persen = model.predict(Xtest20Persen)
print("=====================================================================")
print("Uang pangkal naik 20% dari uang pangkal tahun 2022: Rp.",int(1.2 * UangPangkalTerakhir))
print("Prediksi Jumlah Mahasiswa di jika Uang Pangkal naik 20%:", int(ypred20Persen[0]))
print("=====================================================================")
print()

#uang pangkal 30%
Xtest30Persen = [[2023, 1.3 * UangPangkalTerakhir]]
ypred30Persen = model.predict(Xtest30Persen)
print("=====================================================================")
print("Uang pangkal naik 30% dari uang pangkal tahun 2022: Rp.",int(1.3 * UangPangkalTerakhir))
print("Prediksi Jumlah Mahasiswa di jika Uang Pangkal naik 30%:", int(ypred30Persen[0]))
print("=====================================================================")
print()
```

```python
jumlah_mahasiswa_2022 = data[data['TAHUN'] == 2022]['JUMLAH MAHASISWA'].values[0]
plt.figure(figsize=(10, 6))
plt.plot(data['TAHUN'], data['JUMLAH MAHASISWA'], marker='o', label='Jumlah Mahasiswa')
plt.scatter(2023, ypred10Persen, color='green', label='Prediksi 10%')
plt.plot([2022, 2023], [jumlah_mahasiswa_2022, ypred10Persen], color='green', linestyle='--')
plt.scatter(2023, ypred20Persen, color='blue', label='Prediksi 20%')
plt.plot([2022, 2023], [jumlah_mahasiswa_2022, ypred20Persen], color='blue', linestyle='--')
plt.scatter(2023, ypred30Persen, color='red', label='Predikis 30%')
plt.plot([2022, 2023], [jumlah_mahasiswa_2022, ypred30Persen], color='red', linestyle='--')
plt.title('Prediksi Jumlah Mahasiswa FTI Tahun Ajaran 2023/2024')
plt.xlabel('TAHUN')
plt.ylabel('JUMLAH MAHASISWA')
plt.grid(True)
plt.legend()
plt.show()
```
